## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33 , 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))  # (width, height)

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",  # Title of the Matplot figure
      xlabel="x-axis",  # Title of the x-axis
      ylabel="y-axis")  # Title of the y-axis

# 5. Save & show (you save the whole figure)
fig.savefig("dru_matplotlib.png")
```

### Two options for subplots

```python
# Subplot option 1 (Recommended)
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                           ncols = 2,
                                           figsize=(10,5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.set(title="Nut butter",
       xlabel="Butter",
       ylabel="Price ($)")
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))

# Plot into each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1,].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, it's ok to use the pyplot method
* When plotting something more advanced, use the OO method

```python
over_50 = heart_disease[(heart_disease["age"] > 50)]
over_50.head()
```

```python
# Pyplot method or directly from the pandas dataframe
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10,10));
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,10))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45,80])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Exercise
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"]);

ax.set(title="Heart Disease and Maximum Heart Rate Achieved",
      xlabel="Age",
      ylabel="Max Heart Rate Achieved");

ax.legend(*scatter.legend_elements(), title="Target");

ax.axhline(over_50["thalach"].mean(),
          linestyle=":");
```

```python
# Subplot of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10));

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

ax0.legend(*scatter.legend_elements(), title="Target");

ax0.axhline(over_50["chol"].mean(),
           linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Maximum Heart Rate Achieved",
       xlabel="Age",
       ylabel="Maximum Heart Rate Achieved");

ax1.legend(*scatter.legend_elements(), title="Target");

ax1.axhline(over_50["thalach"].mean(),
           linestyle=":");

# Add title to the figure
fig.suptitle("Heart Disease Analysis",
            fontsize=16,
            fontweight="bold");
```

```python
# Exercise
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10));

# Adding data to ax0
histo = ax0.hist(over_50["chol"])

# Customize ax0
ax0.set(title="Cholesterol Level",
       xlabel="Cholesterol Level")

# Adding data to ax1
histo = ax1.bar(x=over_50["age"],
           height=over_50["thalach"])

# Customize ax1
ax1.set(title="Maximum Heart Rate Achieved",
       xlabel="Age",
       ylabel="Maximum Heart Rate Achieved");

# Add title to the figure
fig.suptitle("Heart Disease Analysis",
            fontsize=16,
            fontweight="bold");
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c=over_50["target"],
                    cmap="winter"); # This changes the colour scheme

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Customizing the y and x axis limitations


# Subplot of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10));

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

ax0.legend(*scatter.legend_elements(), title="Target");

ax0.axhline(over_50["chol"].mean(),
           linestyle="--");

# Change the x axis limits
ax0.set_xlim([50,80])

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

# Customize ax1
ax1.set(title="Heart Disease and Maximum Heart Rate Achieved",
       xlabel="Age",
       ylabel="Maximum Heart Rate Achieved");

ax1.legend(*scatter.legend_elements(), title="Target");

ax1.axhline(over_50["thalach"].mean(),
           linestyle=":");

# Change the x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60, 200])

# Add title to the figure
fig.suptitle("Heart Disease Analysis",
            fontsize=16,
            fontweight="bold");
```
