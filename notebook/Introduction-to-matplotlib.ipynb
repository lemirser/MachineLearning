## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33 , 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))  # (width, height)

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",  # Title of the Matplot figure
      xlabel="x-axis",  # Title of the x-axis
      ylabel="y-axis")  # Title of the y-axis

# 5. Save & show (you save the whole figure)
fig.savefig("dru_matplotlib.png")
```

## Scatter and Bar plot

Making figures with NumPy arrays

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0.0,10.0,num=100).round(2)
x
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x ** 2);
```

```python
# Use the same data and create a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Nut butter prices",
      ylabel = "Prices ($)",
      xlabel = "Butter types");
```

## Histogram and Subplots

```python
# x and y shoud be in a list form
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Subplot option 1 (Recommended)
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                           ncols = 2,
                                           figsize=(10,5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.set(title="Nut butter",
       xlabel="Butter",
       ylabel="Price ($)")
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))

# Plot into each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1,].hist(np.random.randn(1000));
```

### Plotting from Pandas DataFrames

```python
import pandas as pd
```

```python
# Make a DataFrame
car_sales = pd.read_csv("../data/car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2020', periods = 1000))
ts = ts.cumsum()
ts.plot();
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]','')
car_sales
```

```python
# Remove last two zeros
car_sales["Price"] = car_sales["Price"].str[:-2].astype(int)
type(car_sales["Price"][0])
```

```python
# Create a new column for dates
car_sales["Sale Date"] = pd.date_range('1/1/2020', periods=len(car_sales))
car_sales
```

```python
# Create cumulative sum of each car price
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
# Plot car sales
# With pandas you can call the `.plot` api directly at the datafram
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
# Change the type of Price into int
car_sales["Price"] = car_sales["Price"].astype(int)

# Plot the Odometer and price with scatter plot
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

```python
# Plot car sales dataframe using hist
car_sales.plot(x="Make",y="Total Sales",kind="area", figsize=(7,5));
```
